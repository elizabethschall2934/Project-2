```python
tokens = json.loads(r.text)
access_token = tokens['access_token']
access_token
```

```python
get_url = 'https://api.petfinder.com/v2/types'
api_call_headers = {'Authorization': 'Bearer ' + access_token}
filters = {'limit':100}
api_call_types = requests.get(get_url, headers=api_call_headers, params=filters, verify=False)
```

```python
api_call_types.json()
```

```python
animal_types=json.loads(api_call_types.text)
TypeNames = animal_types['types']

for row in TypeNames:
    print(row["name"])
```

```python
animal_type = 'Scales, Fins & Other'
get_url = 'https://api.petfinder.com/v2/animals'
api_call_headers = {'Authorization': 'Bearer ' + access_token}
filters = {'type':animal_type,'limit':100}
api_call_response = requests.get(get_url, headers=api_call_headers, params=filters, verify=False)
```

```python
api_call_response.json()
```

```python
raw_data=json.loads(api_call_response.text)
TotalPages = raw_data['pagination']['total_pages']
TotalCount = raw_data['pagination']['total_count']
print(TotalCount, TotalPages)
```

```python
animal_data = raw_data['animals']
for i in range (2, TotalPages+1, 1):
#for i in range (420, 710, 1):
    filters = {'type':animal_type,'limit':100,'page':i}
    api_call_response = requests.get(get_url, headers=api_call_headers, params=filters, verify=False)
    raw_json = json.loads(api_call_response.text)
    animal_data.extend(raw_json['animals'])
    print(i)
```

```python
len(animal_data)
```

```python
with open(animal_type + '_data.json', 'w', encoding='utf-8') as f:
    json.dump(animal_data, f, ensure_ascii=False, indent=4)
```
